```python
import numpy as np
from matplotlib import pyplot as plt

plt.figure(figsize = (20,8))

s = np.random.normal(0,25,400)
plt.plot(s,label = 'origin')

n = 5     #
rank = []
for i in range(int(len(s)/n)):
    select_s = s[i*n:(i+1)*n]     #
    mid_s = np.median(select_s)
    for j in range(n):       #
        rank.append(mid_s)

plt.plot(rank,label = 'filted')
plt.title('Median filter')
plt.legend()
plt.show() # https://www.bilibili.com/read/cv10858361 bilibili
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.fftpack import rfft, irfft, rfftfreq

fs = 1000
sec = 1
x = np.linspace(0, 1, fs*sec)
y = 5*np.cos(2*np.pi*5*x) + 2.8*np.cos(2*np.pi*x*20)+5.1*np.sin(2*np.pi*60*x)
W = rfftfreq(y.size, d=x[1]-x[0])

plt.plot(x, y)
plt.show()
```

```python
fy = rfft(y)
# fx = np.linspace(0,fs/2,len(fy))
plt.plot(W, fy)
plt.xlim(0,80)
plt.show()

fy[W>30] = 0

ry = irfft(fy)
plt.plot(x, y)
plt.plot(x, ry)
plt.show()
```
